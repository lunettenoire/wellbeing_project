```python
import pandas as pd
df_original = pd.read_csv("world_happiness.csv", encoding="latin-1")
```

```python
# On met df_original au propre: enlever les majuscules et remplacer les espaces par des _
df_original.columns = df_original.columns.str.replace(" ", "_").str.lower()
```

```python
# On transforme ce df_original en csv
df_original.to_csv("df_original.csv", index=False)
```
