```python
import pandas as pd
df_stat_h1 = pd.read_csv("https://raw.githubusercontent.com/lunettenoire/Project/refs/heads/main/df_h1.csv")
df_stat_h1 = df_stat_h1.dropna(subset=["healthy_life_expectancy_at_birth"])
```

**Test de Pearson GDP / Life Ladder**\
_H0 = La corrélation visible entre gdp et Life Ladder est dû au hasard_

```python
from scipy.stats import pearsonr 

# On prend les deux colonnes à comparer
x = df_stat_h1 ["log_gdp_per_capita"]
y = df_stat_h1 ["life_ladder"]

# On execute le test de Pearson
stat, p_value = pearsonr(x, y)

print(f"Coefficient de corrélation (r) : {stat}")
print(f"P-value : {p_value}")

# Interprétation
alpha = 0.05
if p_value < alpha:
    print("Le résultat est statistiquement significatif (on rejette H0)")
else:
    print("Le résultat n'est pas statistiquement significatif (on ne peut pas rejeter H0)")
```
