```python
import pandas as pd
df_h1 = pd.read_csv("df_2015.csv")
```

```python
# On garde uniquement les pays qui ont au moins 7 valeurs renseignées dans la colonne log_gdp_per_capita et on enlève les lignes avec des NaN dans gdp
df_h1 = df_h1[(df_h1.groupby("country_name")["log_gdp_per_capita"].transform("count") >= 7) & (df_h1["log_gdp_per_capita"].notna())]
```

```python
# On exporte en csv
df_h1.to_csv("df_h1.csv", index=False)
```
